```python
#对所有字符串列进行独热编码

%matplotlib inline

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


path="E:/adult.txt"
df2=pd.read_csv(path,names=list("ABCDEFGHIJKLMNO"))

df2.loc[df2.B==' ?','B']=' Private'
df2.loc[df2.G==' ?','G']=' Prof-specialty'

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df2.D=le.fit_transform(df2['D'])
df2.O=le.fit_transform(df2['O'])
df2.J=le.fit_transform(df2['J'])

a=['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']
df2=pd.get_dummies(df2[a]) 
df2
```

```python
all_inputs=df2[df2.columns.drop('O')]
all_classes=df2['O']
input_train,input_test,classes_train,classes_test=train_test_split(all_inputs,all_classes,test_size=0.4,random_state=1)

from sklearn.tree import DecisionTreeClassifier
d_t_c=DecisionTreeClassifier()
d_t_c.fit(input_train,classes_train)
d_t_c.score(input_test,classes_test)
```

```python
%matplotlib inline

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

path="E:/adult.txt"
df=pd.read_csv(path,names=list("ABCDEFGHIJKLMNO"))
df.loc[df.B==' ?','B']=' Private'
df.loc[df.G==' ?','G']=' Prof-specialty'

```

```python
#年龄


labels='<30','30-40','40-50','>50'
colors=['gold','blue','red','green']
explode=(0,0.1,0,0)

plt.pie([4804,5684,4580,3348],explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False,startangle=90)
plt.axis('equal')
plt.show()

```

```python
#种族
white=(df['I']=='White').count()
black=(df.I=='Black').count()
api=(df.I=='Asian-Pac-Islander').count()
aie=(df.I=='Amer-Indian-Eskimo').count()
other=(df.I=='Other').count()


labels='white','black','api','aie','other'
sizes=[white,black,api,aie,other]
colors=['gold','blue','red','yellow','green']
explode=(0,0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False,startangle=90)
plt.axis('equal')
plt.show()

```
